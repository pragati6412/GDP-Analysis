```python
l=df['Country Name'].unique()
list1=[]
for i in range(len(l)):
    df1=df[df['Country Name']==l[i]]
    df2=df1.values
    list1.append(0.0)
    for j in range(1,len(df2)):
        prev=df2[j-1][3]
        curr=df2[j][3]
        x1=float(str(((curr-prev)/prev)*100)[0:5])
        list1.append(x1)
df=df.assign(GDP_growth=list1)
```
