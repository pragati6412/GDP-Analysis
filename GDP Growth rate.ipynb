```python
import pandas as pd
```

```python
df=pd.read_csv('gdp.csv')
df1=df[df['Country Name']=='India']
df2=df[df['Country Name']=='Pakistan']
```

```python
df1
```

```python
df2
```

### For India

```python
d1=df1.values
l=[0]
for i in range(1,len(d1)):
    prev=d1[i-1][3]
    curr=d1[i][3]
    x=float(str(((curr-prev)/prev)*100)[:5])
    l.append(x)
df1=df1.assign(GDP_Growth=l)
df1
```

### For Pakistan

```python
d2=df2.values
l1=[0]
for i in range(1,len(d2)):
    prev=d2[i-1][3]
    curr=d2[i][3]
    x=float(str(((curr-prev)/prev)*100)[:5])
    l1.append(x)
df2=df2.assign(GDP_Growth=l1)
df2
```

```python
df1.groupby('Year').sum()['GDP_Growth'].plot(kind='bar',figsize=(10,6))
```

```python
df1.plot(kind='line',x='Year',y='GDP_Growth',xlabel='YEARS',ylabel='GDP_GROWTH',legend=False,grid=True,figsize=(15,6),
        title='GDP_GROWTH RATE VARIATION YEAR WISE FOR INDIA')
```

```python
df2.plot(kind='line',x='Year',y='GDP_Growth',xlabel='YEARS',ylabel='GDP_GROWTH',legend=False,grid=True,figsize=(15,6),
        title='GDP_GROWTH RATE VARIATION YEAR WISE FOR PAKISTAN')
```
