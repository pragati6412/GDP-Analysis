### For India

```python
df[df['Country Name']=='India']['Value'].describe()
```

```python
df1=df[df['Country Name']=='India']
```

```python
df1.groupby('Year').sum()['Value'].plot(kind='bar',figsize=(10,5))
```

```python
df1.plot(kind='line',x='Year',y='Value',title="India's GDP",figsize=(10,5))
```

### For Arab World

```python
df1=df[df['Country Name']=='Arab World']
```

```python
df1.plot(kind='line',x='Year',y='Value',legend=False,grid=True,xlabel='YEARS',ylabel='GDP',title='Year wise GDP of Arab World',
        figsize=(15,6))
```
